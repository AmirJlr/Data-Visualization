```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.xkcd()
```

```python
epl_df =pd.read_csv('EPL_21.csv',encoding='latin-1')#or encoding='ISO-8859-1'
```

```python
epl_df.info()
```

```python
# finding null values
epl_df.isna().sum()
```

```python
#create 2 new columns

epl_df['MinsPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
```

```python
epl_df.head()
```

```python
#total goals
TotalGoals = epl_df['Goals'].sum()
```

```python
#penalty goals
TotalPenaltyGoals = epl_df['Penalty_Goals'].sum()

#penalty attempts
TotalPenaltyAttempts=epl_df['Penalty_Attempted'].sum()

#missed penalty
MissedPenalty=TotalPenaltyAttempts-TotalPenaltyGoals
```

```python
plt.figure(figsize=(10,6))
data=[TotalPenaltyAttempts,MissedPenalty]
labels=['TotalPenaltyAttempts','MissedPenalty']
color=sns.color_palette('Set2')

plt.pie(data,labels=labels,colors=color,explode=[.1,.1],autopct='%1.1f%%');
```

```python
plt.pie([TotalGoals,TotalPenaltyGoals],labels=['TotalGoals','TotalPenaltyGoals'],explode=[.1,.1],radius=1.5);
```

```python
#Age Disrtibution

under_20 = epl_df[epl_df['Age']<20]
age_20_25 = epl_df[(epl_df['Age'] >= 20) & (epl_df['Age'] < 25)]
age_25_30 = epl_df[(epl_df['Age'] >= 25) & (epl_df['Age'] < 30)]
above30 = epl_df[epl_df['Age']<=30]
```

```python
# Pie Chart for Age
plt.figure(figsize=(12,8))
plt.title('Age Disrtibution',fontsize=25)

data=[len(under_20),len(age_20_25),len(age_25_30),len(above30)]
label=['age < 20','20 <= age < 25','25 <= age < 30','age >= 30']
color=sns.color_palette('Set2')

plt.pie(data,labels=label,colors=color,autopct='%.1f%%');

```

```python
#mean age per club
plt.title('Mean Age Per Club',fontsize=25)
epl_df.groupby('Club')['Age'].mean().plot(kind='bar',figsize=(13,9),color=sns.color_palette('rocket'));
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(x='Club',y='Age',data=epl_df);
plt.xticks(rotation=90);
plt.savefig('box-xkcd.png')
```

```python
most_assist=epl_df.groupby('Club')['Assists'].sum().sort_values(ascending=False)

plt.title('Most Assist per Club',fontsize=25)
most_assist.plot(kind='barh',figsize=(13,9),color=sns.color_palette('magma'));
```

```python
most_goal=epl_df.groupby('Club')['Goals'].sum().sort_values(ascending=False)


plt.title('Most Goal per Club',fontsize=25)
most_goal.plot(kind='barh',figsize=(13,9),color=sns.color_palette('magma'));
```

```python
#top 10 goal

top_goals = epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')

sns.barplot(x='Name',y='Goals',data=top_goals);
plt.xticks(rotation=90);
```

```python
#top 10 assists
top_assists=epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')

sns.barplot(x='Name',y='Assists',data=top_assists);
plt.xticks(rotation=90);
```
